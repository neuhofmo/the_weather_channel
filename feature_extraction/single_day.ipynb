```python
def season_from_month(month):
    season_library = {'winter':[0,0],'spring':[1,0],'summer':[1,1],'autumn':[0,1]}
    
    if month in [12,1,2]:
        monthString = "winter"
    elif month in [3,4,5]:
        monthString = "spring"
    elif month in [6,7,8]:
        monthString = "summer"
    elif month in [9,10,11]:
        monthString = "autumn"
                    
    return season_library[monthString]

month = df_sd_features['Validity date'].apply(lambda x: np.int(x[3:5])).values
df_season = pd.DataFrame(data=np.array(list(map(season_from_month,month))),columns=['season_x','season_y'])
```
